```python
subreddits = ['piratesofthecaribbean', 'nostalgia', 'movies']
query         = 'Dead Men Tell No Tales'
limit         = 10


comments_data = []

for sub_name in subreddits:
    sub = reddit.subreddit(sub_name)
    print(f"\nFetching from r/{sub_name}:")
    for submission in sub.search(query, sort='relevance', time_filter='all', limit=limit):
        print(f"  • {submission.title}  [{submission.id}]")
        # fetch comments
        submission.comments.replace_more(limit=None)
        for c in submission.comments.list():
            comments_data.append({
                'subreddit':    sub_name,
                'post_id':      submission.id,
                'post_title':   submission.title,
                'comment_id':   c.id,
                'comment_time': datetime.fromtimestamp(c.created_utc).strftime('%Y-%m-%d %H:%M:%S'),
                'comment_body': c.body,
                'author':       str(c.author)
            })


df = pd.DataFrame(comments_data)
df.to_csv('pirate5_subreddits_comments.csv', index=False)
print(f"\n✅ Saved {len(df)} comments to 'multi_subreddits_comments.csv'")

```

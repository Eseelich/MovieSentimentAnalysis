```python
# prompt: open a csv from drive/MyDrive/Colab Notebooks/ML/Copy of raw1_vader.csv and preview it and then add a column named "source" and fill in "1" for all the rows

import pandas as pd

file_path = '/content/drive/MyDrive/Colab Notebooks/ML/Copy of raw1_vader.csv'

# Read the CSV file into a pandas DataFrame
try:
  df = pd.read_csv(file_path)
except FileNotFoundError:
  print(f"Error: File not found at {file_path}")
  # You might want to handle this case, perhaps by creating a dummy DataFrame
  # or exiting the program. For now, let's create an empty DataFrame.
  df = pd.DataFrame()

# Preview the first few rows
if not df.empty:
  print("Preview of the DataFrame:")
  print(df.head())

  # Add a new column named "source" with the value 1 for all rows
  df['source'] = 1

  # Preview the DataFrame with the new column
  print("\nDataFrame after adding the 'source' column:")
  print(df.head())
else:
  print("DataFrame is empty.")
```

```python
# prompt: add this to a new df called "vader_all"

vader_all = df.copy()

if not vader_all.empty:
  print("\nPreview of the vader_all DataFrame:")
  print(vader_all.head())
else:
  print("vader_all DataFrame is empty.")
```

```python
# I use the data from WIKIPEDIA to create those plots
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Data of all movies
data = [
    {'Year': 2003, 'Title': 'The Curse of the Black Pearl', 'Franchise': 'Pirates of the Caribbean', 'Budget_m': 140, 'BoxOffice_m': 654.264015},
    {'Year': 2006, 'Title': 'Dead Man’s Chest', 'Franchise': 'Pirates of the Caribbean', 'Budget_m': 225, 'BoxOffice_m': 1066.179725},
    {'Year': 2007, 'Title': 'At World’s End', 'Franchise': 'Pirates of the Caribbean', 'Budget_m': 300, 'BoxOffice_m': 960.996492},
    {'Year': 2008, 'Title': 'Twilight', 'Franchise': 'Twilight-Saga', 'Budget_m': 37, 'BoxOffice_m': 407.187715},
    {'Year': 2009, 'Title': 'The Twilight Saga: New Moon', 'Franchise': 'Twilight-Saga', 'Budget_m': 50, 'BoxOffice_m': 709.711008},
    {'Year': 2010, 'Title': 'The Twilight Saga: Eclipse', 'Franchise': 'Twilight-Saga', 'Budget_m': 68, 'BoxOffice_m': 698.491347},
    {'Year': 2011, 'Title': 'The Twilight Saga: Breaking Dawn – Part 1', 'Franchise': 'Twilight-Saga', 'Budget_m': 127, 'BoxOffice_m': 712.205856},
    {'Year': 2011, 'Title': 'On Stranger Tides', 'Franchise': 'Pirates of the Caribbean', 'Budget_m': 378.5, 'BoxOffice_m': 1045.713802},
    {'Year': 2012, 'Title': 'The Twilight Saga: Breaking Dawn – Part 2', 'Franchise': 'Twilight-Saga', 'Budget_m': 136, 'BoxOffice_m': 829.74682},
    {'Year': 2017, 'Title': 'Dead Men Tell No Tales', 'Franchise': 'Pirates of the Caribbean', 'Budget_m': 275, 'BoxOffice_m': 794.881442},
]



# DataFrame create and sort
df = pd.DataFrame(data).sort_values(['Franchise', 'Year']).reset_index(drop=True)
df['Part'] = df.groupby('Franchise').cumcount() + 1

# Tabelle anzeigen
df
```

```python
# Filter for franchise
twilight = df[df['Franchise'] == 'Twilight-Saga']
pirates = df[df['Franchise'] == 'Pirates of the Caribbean']
```

```python
# Plot for Twilight-Saga
plt.figure()
plt.plot(twilight['Year'], twilight['BoxOffice_m'], marker='o', label='Box Office')
plt.plot(twilight['Year'], twilight['Budget_m'], marker='s', label='Budget')
plt.xticks(twilight['Year'])
plt.xlabel('Jahr')
plt.ylabel('Million USD')
plt.title('Twilight-Saga: Budget vs. Einspielergebnis')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Plot for Pirates of the Caribbean
plt.figure()
plt.plot(pirates['Year'], pirates['BoxOffice_m'], marker='o', label='Box Office')
plt.plot(pirates['Year'], pirates['Budget_m'], marker='s', label='Budget')
plt.xticks(pirates['Year'])
plt.xlabel('Jahr')
plt.ylabel('Million USD')
plt.title('Pirates of the Caribbean: Budget vs. Einspielergebnis')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Plot both movies Box Office over time
plt.figure()
plt.plot(twilight['Year'], twilight['BoxOffice_m'], marker='o', label='Twilight-Saga')
plt.plot(pirates['Year'], pirates['BoxOffice_m'], marker='s', label='Pirates of the Caribbean')
plt.xticks(sorted(df['Year'].unique()))
plt.xlabel('Year')
plt.ylabel('Box Office (Million USD)')
plt.title('Box Office: Twilight vs. Pirates of the Caribbean')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Plot: Box Office by parts
plt.figure()
plt.plot(twilight['Part'], twilight['BoxOffice_m'], marker='o', label='Twilight-Saga')
plt.plot(pirates['Part'], pirates['BoxOffice_m'], marker='s', label='Pirates of the Caribbean')
plt.xticks([1, 2, 3, 4, 5])
plt.xlabel('Part')
plt.ylabel('Box Office (Million USD)')
plt.title('Box Office by Parts')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

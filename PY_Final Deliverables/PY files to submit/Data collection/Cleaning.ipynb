```python
import pandas as pd

#merge all extracted comments
df1 = pd.read_csv('/content/Twilight_relevant_comments.csv')
df2 = pd.read_csv('/content/Twilight_relevant_comments_2.csv')
df3 = pd.read_csv('/content/Twilight_relevant_comments_3.csv')


merged = pd.concat([df1, df2, df3], ignore_index=True)
merged = merged.drop_duplicates()

merged.to_csv('Twilight_merged_comments.csv', index=False)

print(f"✅ Saved {len(merged)} unique rows to 'merged_comments.csv'")

```

```python


dp = pd.read_csv('/content/pirates 5 (1).csv')

print(f" pirate5_merged_cleaned.csv has {dp.shape[0]} rows")

# Group by post_id and take the first title for each
for post_id, group in dp.groupby('post_id', sort=False):
    title = group['post_title'].iloc[0]
    print(f"{post_id}: {title}")
```

```python


#cleaning
df = pd.read_csv('/content/pirate4_subreddits_comments.csv')

df = df.drop_duplicates()

#post ids that don't fit the data
ids_to_extract = [

'1kmdfou', '199azqt', '1cys07g', '1f0iqgr', '6e1pr2', 'wbw8vq', '1jvsj8d', 'xabjx4', '6ev0z4', '6dc1e1', 'ybn0o'
]


mask = df['post_id'].isin(ids_to_extract)
merged_extra   = df[mask].copy()
merged_cleaned = df[~mask].copy()


merged_extra.to_csv('pirate4_merged_extra.csv',   index=False)
merged_cleaned.to_csv('pirate4_merged_cleaned.csv', index=False)

print(f" pirate4_merged_cleaned.csv has {merged_cleaned.shape[0]} rows")

```
